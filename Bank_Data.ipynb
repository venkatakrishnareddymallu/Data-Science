### Bi-variate Analysis

```python
crosstab = pd.crosstab(index=data['Education'], columns=data['Marital'], values=data['Responded'], aggfunc='mean')
seaborn.heatmap(crosstab, annot=True)
```

```python
crosstab = pd.crosstab(index=data['Education'], columns=data['Month'], values=data['Responded'], aggfunc='mean')
seaborn.heatmap(crosstab, annot=True)
```

```python
crosstab = pd.crosstab(index=data['Marital'], columns=data['Month'], values=data['Responded'], aggfunc='mean')
seaborn.heatmap(crosstab, annot=True)
```

```python
crosstab = pd.crosstab(index=data['Education'], columns=data['Job'], values=data['Responded'], aggfunc='mean')
seaborn.heatmap(crosstab, annot=True)
```

```python
crosstab = pd.crosstab(index=data['Education'], columns=data['Age Group'], values=data['Responded'], aggfunc='mean')
seaborn.heatmap(crosstab, annot=True)
```
