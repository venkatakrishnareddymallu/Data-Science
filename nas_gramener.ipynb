```python
#Renaming the Abbreviated states using Dictionary
states={'AN':'Andaman & Nicobar','AP':'Andhra Pradesh','AR':'Arunachal Pradesh','BR':'Bihar','CG':'Chattisgarh','CH':'Chandigarh','DD':'Daman & Diu','DL':'Delhi','DN':'Dadra & Nagar Haveli','GA':'Goa','GJ':'Gujarat','HP':'Himachal Pradesh','HR':'Haryana','JH':'Jharkhand','JK':'Jammu & Kashmir','KA':'Karnataka','KL':'Kerala','MG':'Meghalaya','MH':'Maharashtra','MN':'Manipur','MP':'Madhya Pradesh','MZ':'Mizoram','NG':'Nagaland','OR':'Orissa','PB':'Punjab','PY':'Pondicherry','RJ':'Rajasthan','SK':'Sikkim','TN':'Tamil Nadu','TR':'Tripura','UK':'Uttarakhand','UP':'Uttar Pradesh','WB':'West Bengal'}
```

<h5>REPLACING NUMERICAL DATA WITH CATEGORICAL DATA<h5>

```python
#Using the replace() Method to change type of Data for Better Understanding During Analysis
df['Gender'].replace([1,2],['Boy','Girl'],inplace=True)
df['Age'].replace([2,3,4,5,6,1],['12 years','13 years','14 years','15 years','16+ years','11- years'],inplace=True)
df['Father edu'].replace([5,1,0,2,3,4,],['Degree & above','Illiterate','Not applicable','Primary','Secondary','Sr secondary'],inplace=True)
df['Mother edu'].replace([5,1,0,2,3,4,],['Degree & above','Illiterate','Not applicable','Primary','Secondary','Sr secondary'],inplace=True)
df['Father occupation'].replace([4,0,3,2,8,6,5,7,1],['Clerk','Do not know','Farmer','Labourer','Professional','Business','Skilled Worker','Teacher/Lecturer','Unemployed'],inplace=True)
df['Mother occupation'].replace([4,0,3,2,8,6,5,7,1],['Clerk','Do not know','Farmer','Labourer','Professional','Business','Skilled Worker','Teacher/Lecturer','Unemployed'],inplace=True)
df['Below poverty'].replace([0,1,2],['Dont know','No','Yes'],inplace=True)
df['Use calculator'].replace([1,2],['No','Yes'],inplace=True)
df['Use dictionary'].replace([1,2],['No','Yes'],inplace=True)
df['Read other books'].replace([1,2],['No','Yes'],inplace=True)
df['Help in Study'].replace([1,2],['No','Yes'],inplace=True)
df['Give Lang HW'].replace([2,3,4,1],['1-2 times a week','3-4 times a week','Everyday','Never'],inplace=True)
df['Give Math HW'].replace([2,3,4,1],['1-2 times a week','3-4 times a week','Everyday','Never'],inplace=True)
df['Give Scie HW'].replace([2,3,4,1],['1-2 times a week','3-4 times a week','Everyday','Never'],inplace=True)
df['Give SoSc HW'].replace([2,3,4,1],['1-2 times a week','3-4 times a week','Everyday','Never'],inplace=True)
df['# Books'].replace([3,2,4,1],['11-25 books','1-10 books','25+ books','No books'],inplace=True)
df['Distance'].replace([2,3,4,1],['1-3 km','3-5 km','More than 5 km','Up to 1 km'],inplace=True)
df['Computer use'].replace([5,1,4,3,2],['Daily','No computer','Once in a week','Once in month','Never use'],inplace=True)
df['Library use'].replace([5,1,4,3,2],['More than once in a week','No library','Once in a week','Once or twice in a month','Never use'],inplace=True)
df['Private tuition'].replace([1,2],['No','Yes'],inplace=True)
df['Watch TV'].replace([4,1,2,3],['Every day','Never','Once a month','Once a week'],inplace=True)
df['Read magazine'].replace([4,1,2,3],['Every day','Never','Once a month','Once a week'],inplace=True)
df['Read a book'].replace([4,1,2,3],['Every day','Never','Once a month','Once a week'],inplace=True)
df['Play games'].replace([4,1,2,3],['Every day','Never','Once a month','Once a week'],inplace=True)
df['Help in household'].replace([4,1,2,3],['Every day','Never','Once a month','Once a week'],inplace=True)
df['Siblings'].replace([2,3,4,5],['single child','1 sibling','2siblings','3 siblings'],inplace=True)
df['Correct Lang HW'].replace([2,3,4,1],['1-2 times a week','3-4 times a week','Everyday','Never'],inplace=True)
df['Correct Math HW'].replace([2,3,4,1],['1-2 times a week','3-4 times a week','Everyday','Never'],inplace=True)
df['Correct Scie HW'].replace([2,3,4,1],['1-2 times a week','3-4 times a week','Everyday','Never'],inplace=True)
df['Correct SocS HW'].replace([2,3,4,1],['1-2 times a week','3-4 times a week','Everyday','Never'],inplace=True)
df['Maths is difficult'].replace([3,1,2],['Agree','Disagree','Neither'],inplace=True)
df['Solve Maths in groups'].replace([3,1,2],['Agree','Disagree','Neither'],inplace=True)
df['Solve Maths'].replace([3,1,2],['Agree','Disagree','Neither'],inplace=True)
```

<h1>1.WHAT INFLUENCES STUDENTS PERFORMANCE THE MOST ?<h1>

<h5>1 .TO ANSWER THE QUESTION WE NEED TO AGGREGATE THE DATASET ON MARKS IN DIFFERENT SUBJECTS WITH THE FACTORS(ATTRIBUTES) AND CALCULATE  THE DIFFERENCE  OF THE MAXIMUM AND MINIMUM MARKS BASED ON PARTICULAR FACTOR THAT INFLUENCES CHILD  ON EACH SUBJECT AND VISUALIZE USING HEATMAP,FINALLY MERGE ALL THE INDIVIDUAL MARKS OF SUJBJECT BASED ON PARTICULAR FACTOR(ATTRIBUTE)  <h5>



<h3>CONSIDER GENDER  AS ONE OF THE  FACTOR<h3>

```python
#Using grouop by to aggregate the Dataset on marks with factor
x=df.groupby(['Gender'])['Maths %','Science %','Reading %','Social %','Total'].mean()
#To remove the unfair record other than gender
x.drop(0,inplace=True)
#Calculating the Difference of max and min value that influence the marks based on factor
Gender=pd.DataFrame(x.max()-x.min(),columns=['GENDER'])
# Specifying a subplot for Heatmap
fig,axs=plt.subplots(ncols=2,figsize=(16,5))
plt.figure(figsize=(25,30))
sns.heatmap(data=x,ax=axs[1],annot=True,fmt='.1f')
sns.heatmap(data=Gender,annot=True,fmt='.1f',ax=axs[0])
```

<h6>HERE WE CAN SEE THE GENDER AS FACTOR HEATMAP,WE CAN SEE THAT GENDER INFLUENCE AT 1.9 MARKS IN READING WHICH IS HIGH AND 0.1 MARKS IN SCIENCE WHICH IS LOW AND 0.7 THE AGE INFLUENCE ON THE TOTAL SCORE<h6>
